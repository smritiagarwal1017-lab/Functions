**Quest 1**- What is the difference between a function and a method in Python?

**Answer 1**- In Python, the difference is that a **method** is a function that is associated with an object or class, while a **function** is a standalone block of code that is not tied to any specific object.

All methods are technically functions, but not all functions are methods.

Example of function:
 def say_hello():
    print("Hello!")

say_hello()

Output: Hello!

Example of method:

 class Dog:
    def __init__(self, name):
        self.name = name # attribute
        
    def bark(self): # method
        return f"{self.name} says woof!" # accesses object's data
        
my_dog = Dog("Rex")
print(my_dog.bark())

Output: Rex says woof!

**Quest 5**- What are iterators in Python and how do they differ from iterables?

**Ans 5**- In Python, an **iterable** is an object that can be looped over, while an **iterator** is an object that manages the process of iteration, keeping track of the current state and yielding the next element one at a time. Every iterator is an iterable, but not all iterables are iterators.
Common iterable examples include lists, tuples, strings, and dictionaries whereas iterator is created from an iterable using iter() or as a generator.

**Quest 6**- Explain the concept of generators in Python and how they are defined?

**Ans 6**- In Python, a **generator** is a special type of function that returns a lazy iterator, which produces a sequence of values one at a time, only when requested. This "lazy evaluation" makes them highly memory-efficient, especially for large or infinite datasets, as they don't store the entire sequence in memory

Example: squares = (x * x for x in range(5))

-Using the generator expression:
for num in squares:
    print(num)
Output:
0
1
4
9
16

**Quest 7**- What are the advantages of using generators over regular functions?

**Ans 7**- Generators provide significant advantages over regular functions, primarily through memory efficiency, lazy evaluation, and the ability to handle large or infinite datasets. By yielding values one at a time instead of returning a full list, generators pause execution and resume where they left off, reducing memory overhead and improving performance when dealing with vast data.

**Quest 8**- What is a lambda function in Python and when is it typically used?

**Ans 8**- A lambda function in Python is a small, anonymous function defined using the lambda keyword that can take any number of arguments but is restricted to a single expression. The result of the expression is automatically returned.

Example: # A lambda function to add two numbers
add = lambda x, y: x + y
print(add(2, 3)) # Output: 5

**Quest 10**- What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

-reduce() cumulatively combines elements into a single value.

```python
# Quest 1- Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
# Answer

def even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num

    return even_sum
```

```python
even_numbers([47,11,42,13,22,24])
```

```python
# Question 5- Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
# Answer
class fibonacci:
  def __init__(self,n_terms):
    self.n_terms=n_terms
    self.count=0
    self.a, self.b=0,1
  def __iter__(self):
    return self
  def __next__(self):
    if self.count>=self.n_terms:
      raise StopIteration
    if self.count==0:
      self.count+=1
      return self.a
    elif self.count==1:
      self.count+=1
      return self.b
    else:
      self.a, self.b=self.b,self.a+self.b
      return self.b
```

```python
fib=fibonacci(10)
for num in fib:
  print(num)
```

```python
# Question 6- Write a generator function in Python that yields the powers of 2 up to a given exponent.
# Answer
def power_of_two(n):
  for i in range(n+1):
    yield i**2
gen=power_of_two(10)
gen
```

```python
# Question 7- Implement a generator function that reads a file line by line and yields each line as a string.
# Answer
def read_file(filename):
  with open(filename, 'r') as file:
    for line in file:
      yield line.strip()
    for line in read_file("example.text"):
      print(line)
```

```python
# Question 8- Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
# Answer
t=[(1,6,8),(4,9,6),(2,8,3)]
sorted_t=sorted(t, key=lambda x:x[1])
print(sorted_t)
```

```python
# Question 9- Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# Answer
celsius_temperatures = [25, 30, 15, 20]
fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))
print(fahrenheit_temperatures)
```

```python
# Question 10- Create a Python program that uses `filter()` to remove all the vowels from a given string.
# Answer
def remove_vowels(s):
  vowels = "aeiouAEIOU"
  return "".join(filter(lambda x: x.lower() not in vowels,s))
  string = "Hello, World!"
  print(remove_vowels(string))
```
